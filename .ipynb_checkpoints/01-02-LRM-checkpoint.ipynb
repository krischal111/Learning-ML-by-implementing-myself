## Lets write some Real code for machine learning:
1. A function for using the model to predict the value.
2. A loss function for a particular value
3. A function that sums all the losses.


```python
# General linear_regression
def linear_regression(model:np.array, x:float) -> float:
    y = model[0] 
    for power, coefficient in enumerate(model[1:]):
        y += x**(power+1) * coefficient
    print(model)
    print(x)
    print(y)
    return y

# Very simple linear regression (y = a x)
def simple_linear_regression(model:float, x:float) -> float:
    return model * x

# loss for a single valu
def squared_error(model:float, x:float, y:float, y_predicted = None) -> float:
    if y_predicted is None:
        y_predicted = simple_linear_regression(model, x)

    residue = (y - y_predicted)
    return residue**2

# loss function
def loss_function(model:float, x:np.array, y:np.array, y_predicted = None) -> float:
    if y_predicted is None:
        y_predicted = simple_linear_regression(model, x)

    return np.sum(squared_error(model, x, y, y_predicted))  * (1/2/y.size)

# Check for our use cases
linear_regression(np.array([1]), 5)
linear_regression(np.array([1,2,3]), np.array([3]))
linear_regression(np.array([1,2,3]), np.array([3,2,1]).reshape(-1,1))

print(simple_linear_regression(1, 5))
print(simple_linear_regression(3, np.array([3])))
print(simple_linear_regression(5, np.array([3,2,1])))

```

# Visualization

Codes get messy pretty quickly.
So we need to create a proper visualization to see what are we doing. How things are evolving and changing as we train the model.

```python
# Helps us to get a line of the model
# Uncomment inline version if not working
%matplotlib inline
# %matplotlib widget
import ipywidgets
ipywidgets.IntSlider()

def get_line(model:float, x):
    minx = np.min(x)
    maxx = np.max(x)
    linex = np.array([minx, maxx])
    liney = np.array(simple_linear_regression(model, linex))
    return linex, liney

# Visualize old function (individual images)
def visualize_current(model:float, x, y):
    plt.plot(x,y, 'x')
    linex = np.array([np.min(x), np.max(x)])
    liney = np.array([simple_linear_regression(model, np.min(x)), simple_linear_regression(model, np.max(x))])
    plt.plot(linex, liney)

# Visualize new function (same image)
def ax_visualize_current(line1, model:float, x):
    xx, yy = get_line(model, x)
    line1.set_xdata(xx)
    line1.set_ydata(yy)

model = 5

with plt.ion():
    # Figures and subplot
    figure = plt.figure()
    ax = figure.add_subplot(111)

    # Lines in the plot
    xx, yy = get_line(model, x)
    line1, = ax.plot(xx, yy, label="Predicted Model")
    line2 = ax.scatter(x,y, label="Training Datas")

    # Setting titles and lables on the axes
    plt.title("The datas")
    plt.xlabel("X-values")
    plt.ylabel("Y-values")
    plt.legend()


    # visualize_current(model, x, y)
    # plt.show()
    for _ in range(17):
        ax_visualize_current(line1, model, x)
        figure.canvas.draw()
        figure.canvas.flush_events()
        model *= 0.9
        sleep(0.2)
```

```python
# Actual gradient descent
%matplotlib inline
model = np.random.rand(1)[0] * 40
dg = 0.1
learning_rate = 0.00005

y_predicted = simple_linear_regression(model, x)
loss = loss_function(model, x, y, y_predicted)
for _ in range(25):
    loss2 = loss_function(model + dg, x, y)
    grad_loss = (loss2 - loss)/dg
    model -= learning_rate * grad_loss
    loss = loss_function(model, x, y)
    visualize_current(model, x, y)
    plt.show()
    sleep(0.3)
    print(model, loss, learning_rate)

    dg *= 0.8

print(model)


```
