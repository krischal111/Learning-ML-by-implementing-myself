# MatPlotLib pyplot Cheatsheets

First import main libraries that work well with matplotlib, and pyplot.

MatLab allows us to do calculations on numbers and plot the results very easily.

`numpy` handles the mathematical part.

`pyplot` from the matplotlib handles the mathematical plotting.

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Normal plotting

Generate a linear space with 5 points from 2 to 4 i.e.

$$\begin{array}{ c c c c c}
2.0 & 2.5 & 3.0 & 3.5 & 4.0
\end{array}\\
\begin{array}{|c|c|c|c|}
\ \ \ \ \ &\ \ \ \ \ &\ \ \ \ \ &\ \ \ \ \   \\
\hline\\
\end{array}$$

Then generate some points over the points. Lets say, y = x^3

Then plot it very easily.


```python
x = np.linspace(2, 4, 5) # 

y = (x-3.5)**2

plt.plot(x, y)
```

If plotted for more fine points, you can do so much more.

Let's do the linearly spaced plot on the same interval but with much more number of points as:


Now, you can also add more labels and do so much more stuffs while plotting.

```python
x = np.linspace(2,4,20)
y = (x - 3.5) ** 2
plt.plot(x,y)
```

You can also add more legends, markings and more things as:
# x points


```python
x = np.linspace(2, 4, 20)
y = (x - 3.5) ** 2
fig, ax = plt.subplots()
ax.plot(x, y, color="maroon")
ax.legend()
```
