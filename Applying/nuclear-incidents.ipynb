```python
# for ml
import tensorflow as tf
# for data read
import pandas as pd
# for figure plot
from matplotlib import pyplot as plt
```

```python
# to read the csv file here
df = pd.read_csv("Nuclear Incidents.csv")
```

```python
df[:5]
```

```python
df.Date = pd.to_datetime(df.Date, dayfirst=True)
df.Date[:5]

# sort by date
df.set_index('Date', inplace=True)
df.sort_index(inplace=True)
```

```python
incident_on_year = df.resample('1Y').size()

# this does the main plot
plt.figure(figsize=(15, 5))
incident_on_year.plot(marker='o')
plt.xticks(incident_on_year.index, rotation=0)
plt.grid()
plt.tight_layout()

# plt.xticks(incident_on_year.index, rotation=90)
# plt.grid()
# plt.tight_layout()
# incident_on_year.index
```
