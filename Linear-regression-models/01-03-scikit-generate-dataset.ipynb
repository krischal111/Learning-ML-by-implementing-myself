```python
# Importing libraries
# The following library generates a sparse uncorrelated data, I can generate test data very easily from this
from sklearn.datasets import make_sparse_uncorrelated
# Just for plotting and number related things
from matplotlib import pyplot as plt
import numpy as np

# For stylizing.
from matplotlib import style
style.use("ggplot")
```

```python
# Generating dataset
X, y = make_sparse_uncorrelated(n_samples = 500, n_features = 4);
x = X[:,2]
# X.shape, y.shape
x.shape, y.shape
```

```python
# Plotting the dataset
plt.scatter(x , y)
plt.show
```

```python
# Models: (has parameters), (takes input to get output)
# This case, has only 2 parameters, and takes input with single feature

# Using object oriented approach (instead of functional in prev attempts)
# To represent and use the models.
class Model:
    def __init__(self, parameters = None):
        if parameters is None:
            self.paramerers = np.random.rand((2))
        print(self.paramerers)
    def use_model(self, X):
        sum = self.paramerers[0]
        for (i, a) in enumerate(self.paramerers[1:]):
            sum += a * X **(i+1)
        return sum
    def get_loss_value(self, X, Y):
        y_calc = self.use_model(X)
        residuals = Y - y_calc
        residuals = residuals**2
        return sum(residuals) / (2 * len(residuals))
    def get_plot_points(self, X):
        X = np.array([min(X), max(X)])
        Y = self.use_model(X)
        return X, Y
            
m = Model()
# m.use_model(10)
linex, liney = m.get_plot_points(x)
plt.scatter(x, y)
plt.plot(linex, liney)
plt.show()
loss = m.get_loss_value(x, y)
print(f"Loss = {loss}")

```
